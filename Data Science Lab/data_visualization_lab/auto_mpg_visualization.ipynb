```python
# Load the cleaned dataset into pandas dataframe

import pandas as pd

data = pd.read_csv(r'C:\Users\Nadeem\Documents\datascience\Data Science Lab\data_preprocessing_lab\cleaned_data.csv')

print(data.head())
```

```python
# Install dependency: pip3 install matplotlib --user
# Create a scatter plot using matplotlib

import matplotlib.pyplot as plt

plt.scatter(data['Horsepower'], data['MPG'])
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.title('Horsepower vs MPG')
plt.show()

# Save as an PNG image

plt.savefig('scatter_plot.png')
```

```python
# Install dependency: pip3 install seaborn --user
# Create a histogram using Seaborn

import seaborn as sns

sns.histplot(data['MPG'])
plt.title('Distribution of MPG')
plt.show

# Save as an PNG images

plt.savefig('histogram.png')
```

```python
# Install dependency: pip3 install plotly --user
# Create an interactive plot using plotly

import plotly.express as px

fig = px.scatter(data, x='Weight', y='MPG', title='Weight vs MPG')
fig.show()

# Save as an HTML file

fig.write_html('interactive_plot.html')
```
